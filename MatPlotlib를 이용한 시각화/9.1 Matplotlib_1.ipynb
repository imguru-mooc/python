# Matplotlib를 이용한 시각화

## 데이터를 그래프로 표현하기 1

```python
import pandas as pd

df = pd.read_csv('bok_statistics_CD.csv', header=None)  
print(df.head())
print('\n')

df.columns = ['year', 'CD_rate', 'change']  
df.set_index('year', inplace=True)   
print(df.head())
df.to_csv('bok_statistics_CD_2.csv')
print('\n')

df.plot()

df['CD_rate'].plot()
df['change'].plot()
```

## 그래프를 그림파일로 저장하기
