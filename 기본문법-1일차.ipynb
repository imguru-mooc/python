### 별 출력 하기

```python
# *****
# *****
# *****
# *****
# *****

n = int(input("숫자 입력:"))
for i in range(n):
    for j in range(n):
        print('*', end='')
    print()
```

```python
#     *
#    ***
#   *****
#  *******
# *********

n = int(input("숫자 입력:"))
for i in range(n):
    for j in range(n-i-1):
        print(' ', end='')
    for j in range(i*2+1):
        print('*', end='')
    print()
```

```python
#     *
#    ***
#   *****
#    ***
#     *

n = int(input("숫자 입력:"))
inc = 0
for i in range(n):
    if i <= n//2 :
        for j in range(n//2-inc):
            print(' ', end='')
        for j in range(inc*2+1):
            print('*', end='')
        print()
        inc += 1
    else:
        for j in range(n//2-inc):
            print(' ', end='')
        for j in range(inc*2+1):
            print('*', end='')
        print()
        inc -= 1
```

```python
#     *
#    ***
#   *****
#    ***
#     *

n = int(input("숫자 입력:"))
inc = 0
sign = 1
for i in range(n):
    for j in range(n//2-inc):
        print(' ', end='')
    for j in range(inc*2+1):
        print('*', end='')
    print()
    
    if i == n//2 :
        sign = -sign
        
    inc += sign
```

```python
# *   *
# ** **
# *****
# ** **
# *   *

n = int(input("숫자 입력:"))
inc = 0
sign = 1
for i in range(n):
    for j in range(inc+1):
        print('*', end='')
    for j in range(n-(inc+1)*2):
        print(' ', end='')
    for j in range(inc+1):
        print('*', end='')
    print()
    
    if i == n//2 :
        sign = -sign
        
    inc += sign
```
